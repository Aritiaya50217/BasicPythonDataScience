```python
# Dictionary

import pandas as pd


data = { 'Col1' : [110 , 111 , 112 , 113 ],
           'Col2' : [210 , 211 , 212 , 213],
           'Col3' : [310 , 311, 312, 313] }

print(type(data))

df = pd.DataFrame(data,index = ['p1','p2','p3','p4'] )

display(df)
```

```python
# Value ของแต่ละคอลัมน์ สามารถนำวิธีการต่าง ๆ ของ Numpy มาสร้างข้อมูลในแบบอาร์เรย์ได้

import pandas as pd
import numpy as np

data = { 'Mon' : np.arange(10,14) ,
            'Wed' : np.random.randint(10,100,4) ,
            'Fri' : np.full(4,10) }
df = pd.DataFrame(data,index = ['p1','p2','p3','p4'] )
display(df)
```

```python
# การนำข้อมูล dictionary หลาย ๆอันมารวมกัน เป็นข้อมูลของ Object DataFrame

import pandas as pd
import numpy as np

a = {'p1' : 200,'p2' : 300,'p3' : 300,'p4' : 333}
b = {'p1' : 20,'p2' : 30,'p3' : 5}
c = {'p1' : 20,'p2' : 30,'p4' : 5}

data  ={'price' : a , 'sales' : b , 'stock' : c }

df = pd.DataFrame(data,dtype = object)
display(df)
```

```python
# การนำข้อมูลมารวมกันใน DataFrame  สามารถเป็นได้หลายแบบ เช่น ลิสต์ อาร์เรย์ หรือ ทูเพิล
import pandas as pd
import numpy as np

# ข้อมูลแบบกำหนดคอลัมน์  price เก็บข้อมูลแบบ list  ,  sales เก็บข้อมูลแบบอาร์เรย์ , stock เก็บข้อมูลแบบทูเพิล
data = {'price' : [200, 300, 345, 500] ,
           'sales' : np.array([0, 30, 34, 89]),
           'stock' : (8,7,7,7) }

df = pd.DataFrame(data , index = ['p1', 'p2', 'p3' ,'p4'] )
display(df)
```
