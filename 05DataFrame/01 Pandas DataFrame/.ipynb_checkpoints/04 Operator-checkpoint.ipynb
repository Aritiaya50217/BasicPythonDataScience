```python
# การเลือกข้อมูลโดยใช้โอเปอร์เรเตอร์ at , iat
# การเลือกข้อมูลใน DataFrame  โดยวิธีระบุแถวและคอลัมน์ ทำได้โดยการใช้ at , iat 
# df.at[แถว , คอลัมน์]  หรือ df.iat[ลำดับแถว , ลำดับคอลัมน์]

import pandas as pd
import numpy as np

data = { 'price' : [222,4447,888,999] ,
           'sales'  : [4,68,9,9],
           'stock'  : [9,8,6,5],
           'discount' : [5,4,5,6]  }

df = pd.DataFrame(data,index = ['p1' ,'p2','p3','p4'])
display(df)

# at ใช้เมื่อเราต้องการระบุแถวและคอลัมน์เป้าหมายด้วย ชื่ออ้างอิง 

display(df.at['p1','price'],df.at['p4' , 'sales'])

# iat ใช้เมื่เราต้องระบุแถวและคอลัมน์เป้าหมายด้วยเลขลำดับ
display(df.iat[0,0],df.iat[3,3])
```

```python
# Index Slicing
# df [start : stop : step]

import pandas as pd
import numpy as np

data = { 'One' : np.random.randint(100,1000,10),
            'Two' : np.random.uniform(1,10,10),
            'Three' : np.random.rand(10) ,
            'Four' : np.random.randint(1,100,10)  }

df = pd.DataFrame(data , index = list('ABCDEFGHIJ'))

display(df)
print("ข้อมูลแถวที่ 0 - 5 : ")
display(df[:5])

print("ข้อมูลแถวที่ 5 : ")
display(df[5:6])
```

```python
# Fancy Index  + Index Slicing 
# รูปแบบ  df [['col1' , 'col2' , 'col3' ,...]] [start : stop : step]

import pandas as pd
import numpy as np

data = { 'One' : np.random.randint(100,1000,10),
            'Two' : np.random.uniform(1,10,10),
            'Three' : np.random.rand(10) ,
            'Four' : np.random.randint(1,100,10),
            'Five' : np.random.uniform(10,50,10) ,
            'Six' : np.random.choice([101 , 555 , 108 , 123, 777] , 10)}

df = pd.DataFrame(data , index = list('ABCDEFGHIJ'))
display(df)

# df [['col1' , 'col2' , 'col3' ,...]] [start : stop : step]
print('\nแสดงข้อมูลคอลัมน์ One Three Five ระหว่าง F ถึง I :')
display(df[['One' ,'Three' ,'Five']] ['F': 'I'])

# df [start : stop : step] [['col1' , 'col2' , 'col3' ,...]] 
print('\nแสดงข้อมูลคอลัมน์ One Three Five ของลำดับสุดท้าย:')
display(df[-1 : ] [['One' ,'Three' ,'Five']] )

print('\nเลือกข้อมูลตั้งแต่ลำดับแรก เว้นระยะ 3 ลำดับ : ')
display(df[0::3 ] )
```

```python
#  การใช้ loc และ iloc ร่วมกับ Fancy Index

import pandas as pd
import numpy as np

data =  { 'One' : np.random.randint(100,1000,10),
            'Two' : np.random.uniform(1,10,10),
            'Three' : np.random.rand(10) ,
            'Four' : np.random.randint(1,100,10)  }

df = pd.DataFrame(data,index=list('ABCDEFGHIJ'))
display(df)

# หากต้องการเลือกเฉพาะบางแถวของทุกคอลัมน์ โดยระบุค่าแบบ Label Index ใช้ loc
# df.loc[ ['label1' ,'label2','label3',...]]
print('\nLabel Index แสดงข้อมูลของลำดับ A E G I : ')
display(df.loc[['A','E','G','I']])

# หากต้องการเลือกเฉพาะบางแถวของทุกคอลัมน์ โดยระบุค่าแบบ Position Index ใช้ iloc
# df.iloc[ ['pos1' ,'pos2','pos3',...]]

print('\nPosition Index แสดงข้อมูลของลำดับที่ 1 3 5 9 : ')
display(df.iloc[[1,3,5,9]])
```
