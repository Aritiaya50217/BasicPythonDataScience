```python
# การกำหนดข้อมูลให้กับ dataFrame

# รูปแบบการใช้งาน  pd.dataFrame(data,index,columns,dtype)

# data ข้อมูลที่จัดเก็บใน DataFrame
# index คือลำดับแถว
# columns ชื่อของแต่ละคอลัมน์
# dtype คือ ชนิดข้อมูลที่จัดเก็บในแต่ละคอล้มน์

import pandas as pd

data = [[10,20,60,80],
           [5,6,7,9],
           [12,44,55,66],
           [1,5,5,5,5],
           [1,2,3,4,5]]

df = pd.DataFrame(data,index=list('ABCDE'),columns=['One','Two','Three','Four','Five'])

display(df)
```

```python
import pandas as pd
import numpy as np

data1 = np.array([[11,22,33,44,55],
                            [33,66,77,8,9],
                            [0,8,8,7,7]])

# นำข้อมูลใน data1 มาใส่ในรูปแบบของ DataFrame
df1 = pd.DataFrame(data1,columns=list('ABCDE'))

data2 = np.random.uniform(1,11,(4,5))

df2 = pd.DataFrame(data2,columns =['one','two','three','four','five'] )
display(df1,df2)
```

```python
import pandas as pd
import numpy as np

data1 = np.array([[11,22,33,44,55],
                            [33,66,77,8,9],
                            [0,8,8,7,7]])

# นำข้อมูลใน data1 มาใส่ในรูปแบบของ DataFrame
df1 = pd.DataFrame(data1,columns=list('ABCDE'))

data2 = np.random.uniform(1,11,(4,5))
# กำหนดชื่อ index
df2 = pd.DataFrame(data2,columns =['one','two','three','four','five'],index=['idx1','idx2','idx3','idx4'])
display(df1,df2)
```
