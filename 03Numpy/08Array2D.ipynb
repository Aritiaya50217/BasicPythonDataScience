```python
# array()  อาจกำหนดสมาชิกลงไปโดยตรงหรือจะสร้างเป็นลิสต์ 2 มิติลงไปก็ได้
import numpy as np

# กำหนดสมาชิกลงไปโดยตรง
a1 = np.array([[10,20,30,40],[50,60,70,80],[11,22,33,44],[55,66,77,88]])
# สร้างเป็นลิสต์
list2d = [[10,20,30,40],[50,60,70,80]]

print(a1)
print(list2d)

# นำลิสต์ที่สร้างใส่ไปใน array
a2 = np.array(list2d)
a2
```

```python
# zeros([rows,cols])  ระบุจำนวนแถวและคอลัมน์ สมาชิกทุกตัวมีค่าริ่มต้นเป็น 0
# ones([rows,cols]) ระบุจำนวนแถวและคอลัมน์ สมาชิกทุกตัวมีค่าริ่มต้นเป็น 1
# full([rows,cols],value)  ระบุจำนวนแถวและคอลัมน์ พร้อมกับค่าเริ่มต้นที่ต้องการ

import numpy as np
a0 = np.zeros([3,4] ,dtype=int)
a1 = np.ones([4,6],dtype = int)
af = np.full([4,4],10)

print(a0,a1,af, sep='\n\n')
```

```python
# random.randint(low, high, [rows,cols])  สร้างอาร์เรย์ของเลขจำนวนเต็มภายในช่วงที่ระบุ
#  random.uniform(low, high, [rows,cols])  สร้างอาร์เรย์ของเลขทศนิยมภายในช่วงที่ระบุ
# random.rand(rows , cols)  สร้างอาร์เรย์ที่เป็นเลขทศนิยมระหว่าง 0.0 - 1.0 
# random.choice([sequence] , [rows,cols])  สุ่มค่าจากรายการตัวเลือกที่กำหนด พร้อมระบุจำนวนแถวและคอลัมน์

import numpy as np

# สร้างอาร์เรย์แบบสุ่มตัวเลขจำนวนเต็มตั้งแต่ 10 - 101 มา 5 แถว 5 คอลัมน์
rand_int  = np.random.randint(10,101,[5,5])

# สร้างอาร์เรย์แบบสุ่มตัวเลขทศนิยมตั้งแต่ 1 - 11 มา 4 แถว 5 คอลัมน์
uniform = np.random.uniform(1,11,(4,5))

#  สร้างอาร์เรย์ที่เป็นเลขทศนิยมระหว่าง 0.0 - 1.0 มา 6 แถว 5 คอลัมน์
rand = np.random.rand(6,5)

# สุ่มค่าจากรายการตัวเลือก ดังนี้ 3.114,0.345,-10,9.99,33,-2.57  มา 3 แถว 4 คอลัมน์
choice = np.random.choice([3.114,0.345,-10,9.99,33,-2.57],(3,4))

print(rand_int,uniform,rand,choice, sep='\n\n')
```

```python
# การเข้าถึงสมาชิกของอาร์เรย์ 2 มิติ
# array[ลำดับแถว , ลำดับคอลัมน์]
import numpy as np

a = np.random.randint(1,11,[3,4])

print(a,'\n')
print('แถวที่ 1 :',a[1])
print('แถวที่ 2 :',a[2],'\n')
print('แถวที่ 1 คอลัมที่ 2 คือ :',a[1][2])
print('แถวที่ 0 คอลัมที่ 3 คือ :',a[0][3])
```

```python
# หากต้องการเข้าถึงสมาชิกที่เป็นลำดับต่อเนื่องกัน สามารถใช้วิธี index slicing ได้
# รูปแบบ array[start : stop : step , start : stop : step ]   กลุ่มแรกคือลำดับแถว กลุ่มหลังคือลำดับคอลัมน์

import numpy as np
a = np.random.randint(1,11,[6,6])
print(a,'\n')
print('ข้อมูลช่วง แถวที่ 1 - 4 และ คอลัมน์ที่ 2 - 3 : \n',a[1:5, 2:4],'\n')
print('ข้อมูลช่วง แถวที่ 0 - 3 และคอลัมน์ที่ 2 - 3 :\n',a[:4,2:],'\n')
print('เลือกทุกแถว และ คอลัมน์ที่ 0-5 โดยมีระยะห่างระหว่างข้อลัมน์ เท่ากับ 2 :\n',a[:: , :6:2])
```
